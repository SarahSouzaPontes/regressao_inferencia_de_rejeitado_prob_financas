Separar em variaveis preditivas e respostas

```python
X = df.drop('MonthlyIncome', axis=1)
y = df.MonthlyIncome
#var resp MonthlyIncome
```

```python
X.shape
#apenas var resposta MonthlyIncome = y
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_tr, y_ts = train_test_split(X,y, test_size=.3)
#separando parte que o modelo nunca viu para ver o treinamento
```

TREINAR MODELO

```python
from sklearn.ensemble import RandomForestRegressor
#arvores de decisao -> regressor
rf = RandomForestRegressor(n_estimators=500, max_depth=8, n_jobs=-1)
rf.fit(X_tr, y_tr)
#var de treino e teste (preditiva e de resposta)
```

```python
y_ts
#linha e valor
```

```python
rf.predict(X_ts)
#predição não é mais predict proba como classificação e apenas predicao
#4363.10745772, -> RENDA REAL  6169.0
#2802.93026575, 
#6664.0691456 , ..., 
#5874.93481451,
#4777.90034667, 
#9208.60446197
```

```python
erro = y_ts - rf.predict(X_ts)
erro
```

```python
erro.mean()
#RENDA REAL MENOS  predição
```

```python
erro_abs = np.abs(y_ts - rf.predict(X_ts))
erro_abs
#erro absoluto -> apenas serão positivos não importa se para cima ou para baixo
#quanto o modelo errou
```

```python
erro_abs.mean()
#media do erro, medida mais real
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_ts, rf.predict(X_ts))
#metrica pronta para isso é erro absoluto médio, calculando direto
```

```python
#Modelo há vies?
#comprara histogramas
plt.figure(figsize=(15,5))
plt.hist(y_ts,             bins=np.logspace(1,6.5,100), color='b', alpha=.30, density=True)
#histograma da realidade
plt.hist(rf.predict(X_ts), bins=np.logspace(1,6.5,100), color='r', alpha=.30, density=True)
#Histograma da predição
plt.xscale('log')
```
